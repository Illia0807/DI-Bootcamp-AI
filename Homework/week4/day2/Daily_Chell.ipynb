```python
import numpy as np
import pandas as pd
```

```python
data = np.random.uniform(-5, 35, size=(10, 12))
print(data)

cities = ['New York', 'Tokyo', 'Paris', 'London', 'Berlin',
          'Sydney', 'Toronto', 'Dubai', 'Moscow', 'S찾o Paulo']

months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
df = pd.DataFrame(data, index=cities, columns=months)
print(df)
```

```python
annual_avg = df.mean(axis=1)
hottest_city = annual_avg.idxmax()
coldest_city = annual_avg.idxmin()
print('hotest city '+hottest_city)
print('coldest cit '+coldest_city)
```

```python
import matplotlib.pyplot as plt
df.T.plot(figsize=(10, 6), title='Monthly Temperature Trends')
plt.xlabel('Month')
plt.ylabel('Temperature (째C)')
plt.show()
```
