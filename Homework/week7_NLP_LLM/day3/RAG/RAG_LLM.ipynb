# code for vector database and embedding functions

```python
from sentence_transformers import SentenceTransformer # Импортируем для создания эмбеддингов в Colab

# Имитируем модели, как в туториале.
# В реальном локальном Ollama они бы использовались.
EMBEDDING_MODEL_NAME = 'sentence-transformers/all-MiniLM-L6-v2' # Используем популярную модель для эмбеддингов из HF
LANGUAGE_MODEL_NAME = 'bartowski/Llama-3.2-1B-Instruct-GGUF' # Это имя модели для Ollama, не используется напрямую в Colab для генерации

# Загружаем модель для создания эмбеддингов в Colab
embedding_model = SentenceTransformer(EMBEDDING_MODEL_NAME)

# Каждая запись в VECTOR_DB будет кортежем (chunk, embedding)
VECTOR_DB = []

# Имитация функции ollama.embed с использованием sentence_transformers
# В реальном локальном Ollama ты бы использовал ollama.embed напрямую
def add_chunk_to_database(chunk):
  # Создаем эмбеддинг с помощью sentence_transformers
  # .tolist() преобразует тензор PyTorch в обычный список Python
  embedding = embedding_model.encode(chunk).tolist()
  VECTOR_DB.append((chunk, embedding))

print("Начинаем создание векторной базы данных...")
for i, chunk in enumerate(dataset):
  add_chunk_to_database(chunk)
  # Выводим прогресс каждые 10 чанков, чтобы не заспамить вывод
  if (i + 1) % 10 == 0 or (i + 1) == len(dataset):
    print(f'Добавлено чанков: {i+1}/{len(dataset)}')

print(f'Векторная база данных создана. Всего {len(VECTOR_DB)} записей.')
print(f'Размерность эмбеддинга: {len(VECTOR_DB[0][1])}') # Покажем размерность эмбеддингов
```

# Implementing the Retrieval Function
use cosine similarity

```python
import numpy as np # Обычно np импортируют для числовых операций

# Функция для вычисления косинусного сходства между двумя векторами
def cosine_similarity(a, b):
  # Переводим в numpy массивы, чтобы использовать np.dot и np.linalg.norm для эффективности
  # Если векторы уже numpy массивы, то преобразование не нужно
  a = np.array(a)
  b = np.array(b)
  dot_product = np.dot(a, b)
  norm_a = np.linalg.norm(a)
  norm_b = np.linalg.norm(b)
  # Избегаем деления на ноль, если один из векторов нулевой (что маловероятно для эмбеддингов)
  if norm_a == 0 or norm_b == 0:
      return 0.0
  return dot_product / (norm_a * norm_b)

# Функция для поиска релевантных чанков
def retrieve(query, top_n=3):
  # Создаем эмбеддинг для запроса пользователя
  # Используем ту же модель, что и для создания VECTOR_DB
  query_embedding = embedding_model.encode(query).tolist() # Используем embedding_model из предыдущего шага

  # Временный список для хранения пар (чанк, сходство)
  similarities = []
  for chunk, embedding in VECTOR_DB:
    similarity = cosine_similarity(query_embedding, embedding)
    similarities.append((chunk, similarity))

  # Сортируем по сходству в убывающем порядке (чем выше сходство, тем релевантнее чанк)
  similarities.sort(key=lambda x: x[1], reverse=True)

  # Возвращаем N наиболее релевантных чанков
  return similarities[:top_n]

print("Функции 'cosine_similarity' и 'retrieve' определены.")

# Простой тест функции retrieve
test_query = "How fast can a cat run?"
print(f"\nТестовый запрос: '{test_query}'")
retrieved_results = retrieve(test_query, top_n=2)
print("Найденные релевантные факты:")
for chunk, similarity in retrieved_results:
    print(f" - (Сходство: {similarity:.4f}) {chunk.strip()}")
```
